##### Data cleaning and preparation

```python
df.isnull().sum()
```

```python
mode=df[['LB','AC','DS','DP','MLTV','Width','Tendency','NSP']].mode()
df1=df.fillna(mode)
df1
```

```python
df1[df1.duplicated()]
```

```python
df1=df1.drop_duplicates()
df1
```

```python
df2=[i for i in df1.columns if df1[i].dtypes!='0']
df2
```

```python
df1.describe()
```

#### Calculation of outliers

```python
Q1=df1['FM'].quantile(0.25)
Q1
```

```python
Q3=df1['FM'].quantile(0.75)
Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
lower_quantile=Q1-1.5*IQR
lower_quantile
```

```python
upper_quantile=Q3+1.5*IQR
upper_quantile
```

```python
df1_no_outliers=df1[(df['FM']>lower_quantile)&(df1['FM']<upper_quantile)]
df1_no_outliers
```

```python
# Boxplot
plt.figure(figsize=(10,15))
df1_no_outliers.boxplot()
plt.show()
```
